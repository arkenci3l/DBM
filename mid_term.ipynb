```python
import sqlite3

print(sqlite3.version)
print(sqlite3.sqlite_version)
```

```python
db = sqlite3.connect(':memory:')
```

```python
cursor = db.cursor()
```

```python
cursor.execute('''CREATE TABLE Countries
(
  country_id CHAR(13) NOT NULL PRIMARY KEY, 
  languages_spoken VARCHAR2(20) NOT NULL, 
  cname VARCHAR2(20) NULL
  )''')
db.commit()
```

```python
cursor.execute('''INSERT INTO Countries values ("ISO 3166-2:AF" , 'Dari & Pashto', 'Afghanistan')
''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE Drivers
(
  license_number     CHAR(18)  NOT NULL PRIMARY KEY,
	dname              VARCHAR(20)  NULL ,
	clearance_level    NUMBER(4)  NULL ,
	dlanguage_spoken   VARCHAR2(20)  NOT NULL
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE drops_off_location
(
  drops_off_location_name VARCHAR2(20)  NOT NULL PRIMARY KEY,
	adress             VARCHAR(20)  NULL ,
	street_number      INTEGER  NULL ,
  city               VARCHAR2(20)  NOT NULL 
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE FATL
(
  license_number     CHAR(18)  NOT NULL PRIMARY KEY,
	level              NUMBER(2)  NOT NULL ,
	qualification_date DATETIME  NOT NULL ,
  FOREIGN KEY (license_number) REFERENCES Drivers(license_number)
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE Officials
(
  City_ID            CHAR(8)  NOT NULL ,
	oname             VARCHAR(20)  NULL ,
	role               VARCHAR(20)  NULL ,
	preferred_spoken_language VARCHAR2(10)  NOT NULL ,
	official_id       VARCHAR2(20)  NOT NULL PRIMARY KEY,
	country_id         CHAR(13)  NULL ,
  FOREIGN KEY (country_id) REFERENCES Countries(country_id) ON DELETE SET NULL
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE Out_of_service_vehicules
(
  odometer           NUMBER  NOT NULL ,
	final_cost         INTEGER  NOT NULL ,
	description_of_actions VARCHAR2(60)  NOT NULL ,
	vdate               DATETIME  NOT NULL ,
	action_type        CHAR(1)  NOT NULL ,
	vehicule_id        CHAR(5)  NOT NULL PRIMARY KEY , 
  FOREIGN KEY (vehicule_id) REFERENCES Vehicules(vehicule_id)
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE pickup_location
(
  pickup_location_name VARCHAR2(20)  NOT NULL PRIMARY KEY,
	adress             VARCHAR(20)  NULL ,
	street_number      INTEGER  NULL ,
	city               VARCHAR2(20)  NOT NULL 
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE STLVT
(
  license_number     CHAR(18)  NOT NULL PRIMARY KEY,
	level              NUMBER(1)  NOT NULL ,
	qualiffication_date DATETIME  NOT NULL ,
	certifying_authority VARCHAR2(30)  NOT NULL ,
  FOREIGN KEY (license_number) REFERENCES Drivers(license_number) 
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE travel
(
  booking_reference_number NUMBER  NOT NULL PRIMARY KEY,
	start_date_and_time DATETIME  NOT NULL ,
	end_date_and_time  DATETIME  NOT NULL ,
	start_odometer_value INTEGER  NULL ,
	end_odometer_value INTEGER  NULL ,
	pickup_location_name VARCHAR2(20)  NULL ,
  drops_off_location_name VARCHAR2(20)  NOT NULL ,
  FOREIGN KEY (pickup_location_name) REFERENCES pickup_location(pickup_location_name) ON DELETE SET NULL ,
  FOREIGN KEY (drops_off_location_name) REFERENCES drops_off_location(drops_off_location_name) ON DELETE SET NULL 
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE Vehicules
(
  vehicule_id        CHAR(5)  NOT NULL PRIMARY KEY,
	registration_Id    CHAR(8)  NOT NULL ,
	Manufacturer       VARCHAR(20)  NULL ,
	model              VARCHAR(20)  NULL ,
	color              VARCHAR(10)  NULL ,
	odometer           NUMBER  NOT NULL ,
	passenger_capacity INTEGER  NOT NULL 
  )''')
db.commit()
```

```python
cursor.execute('''CREATE TABLE Driver_booking_service
(
  vehicule_id CHAR(5) NOT NULL ,
	official_id       VARCHAR2(20)  NOT NULL ,
	license_number     CHAR(18)  NOT NULL ,
	matching_language  VARCHAR2(20)  NOT NULL ,
	booking_reference_number NUMBER  NOT NULL PRIMARY KEY , 
  FOREIGN KEY (vehicule_id) REFERENCES Vehicules(vehicule_id) ,
  FOREIGN KEY (official_id) REFERENCES Officials(official_id) , 
  FOREIGN KEY (license_number) REFERENCES Drivers(license_number)
  FOREIGN KEY (booking_reference_number) REFERENCES travel(booking_reference_number)
  )''')
db.commit()
```
